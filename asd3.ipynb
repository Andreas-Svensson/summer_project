```python
import sounddevice as sd
import plotly.express as px
import time
import numpy as np
from Scripts.wave_test import generate_waveform
```

```python
f = 440
sine_440 = generate_waveform(f, 'sine', duration=0.01)
square_440 = generate_waveform(f, 'square', duration=0.01)
```

```python
import pandas as pd

# x-axis time array
t = np.linspace(0, 0.01, len(sine_440))

df = pd.DataFrame({'Time': t, 'sine_440': sine_440, 'square_440': square_440})

fig = px.line(df, x='Time', y=['sine_440', 'square_440'])

fig.update_layout(
    xaxis_title='Time (ms)',
    yaxis_title='Amplitude',
    title='Normalized sine and square wave @ 440Hz'
)
```

```python
from Scripts.wave_test import transpose_waveform

t = np.linspace(0, 0.01, len(sine_440))

df = pd.DataFrame({'Time': t, 'sine_440': sine_440, 'transposed': np.resize(transpose_waveform(sine_440, 4, 44100), len(sine_440))})

fig = px.line(df, x='Time', y=['sine_440', 'transposed'])

fig.update_layout(
    xaxis_title='Time (ms)',
    yaxis_title='Amplitude',
    title='Normalized sine @ 440hz and transposed by a factor of 4 (then resized by np - this is not correct)'
)
```
